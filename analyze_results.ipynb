```python
import pandas as pd
import re
import numpy as np
```

```python
real_times_parallel = []

# Open the file in read mode
with open("pcd_1.txt", 'r') as file:
    for line in file:
        # Check if the line contains 'real'
        if line.startswith("real"):
            # Use regular expression to extract the number
            match = re.search(r'(\d+m\d+\.\d+s)', line)
            if match:
                # Extract the time value (in format '0mX.XXXs')
                parsed_time = float(match.group(1)[2:-1])
                real_times_parallel.append(parsed_time)
                
real_times_parallel
```

```python
real_times_sequential = []

# Open the file in read mode
with open("pcd_2.txt", 'r') as file:
    for line in file:
        # Check if the line contains 'real'
        if line.startswith("real"):
            # Use regular expression to extract the number
            match = re.search(r'(\d+m\d+\.\d+s)', line)
            if match:
                # Extract the time value (in format '0mX.XXXs')
                parsed_time = float(match.group(1)[2:-1])
                real_times_sequential.append(parsed_time)
                
real_times_sequential
```

```python
mean_value_sequential = sum(real_times_sequential) / len(real_times_sequential)
mean_value_parallel = sum(real_times_parallel) / len(real_times_parallel)
mean_value_sequential, mean_value_parallel
```

```python
import matplotlib.pyplot as plt

# Example data
categories = ['Sem concorrência', 'Com concorrência']
values = [1000*mean_value_sequential, 1000*mean_value_parallel]

# Specify colors for each bar
colors = ['blue', 'green']
# Create a bar chart
bars = plt.bar(categories, values, color=colors)

# Add title and labels
#plt.title('Bar Chart with 2 Values')
plt.ylabel('Tempo médio de execução (ms)')
plt.xlabel('Versão do programa')

for bar in bars:
    # Get the height of the bar
    yval = bar.get_height()
    # Place the value label just above the bar
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, round(yval, 2), 
             ha='center', va='bottom', fontsize=12)


max_value = max(values)

# Add extra space to the top of the bars (set y-axis limit)
plt.ylim(0, 1.1*max_value )  # Add 1 unit above the maximum value

# Show the plot
plt.show()
```
